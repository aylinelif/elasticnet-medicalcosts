```python
import pandas as pd 
df = pd.read_csv("insurance.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = pd.get_dummies(df,columns = [ "sex" , "smoker" , "region" ],drop_first=True)
```

```python
#target 
y = df["charges"]
#features
X = df.drop("charges",axis=1 )
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.40,random_state=42)
```

```python
#standartScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
model = ElasticNet()
model.fit(X_train_scaled,y_train)
```

```python
#Measure performance
import numpy as np
from sklearn.metrics import r2_score,mean_squared_error
y_pred = model.predict(X_test_scaled)
r2 = r2_score(y_test,y_pred)
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print("R2: ",r2)
print("RMSE: ",rmse)
```

```python
from sklearn.linear_model import ElasticNetCV
e_net_cv = ElasticNetCV(l1_ratio=[0.1, 0.3, 0.5, 0.7, 0.9],
    alphas=[0.001, 0.01, 0.1, 1, 10],cv =5,random_state=42)
```

```python
e_net_cv.fit(X_train_scaled,y_train)
```

```python
print(" best alpha : ",e_net_cv.alpha_)
```

```python
print(" best l1_ratio : ",e_net_cv.l1_ratio_)
```

```python
y_pred = e_net_cv.predict(X_test_scaled)

print("Final R2:", r2_score(y_test, y_pred))
print("Final RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
```
